```python
# Importe as bibliotecas

import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModel

# Passo 2: Tokenização

# escolher a frase
frase = "O mercado financeiro está estável"

# Carregue o tokenizador BERT
bert_modelo = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(bert_modelo)

#Converta a frase em tokens:
tokens = tokenizer(frase, return_tensors="tf")
print(tokens)
```

```python
# sugerir outra frase

frase_teste = "A minha gata comeu a minha bola"

tokens2 = tokenizer(frase_teste, return_tensors="tf")
print(tokens2)
```

```python
# Passo 3: Testando o Mecanismo de Atenção (Self-Attention)

# Carregar o modelo pré-treinado BERT
modelo = TFAutoModel.from_pretrained(bert_modelo)

# Obter a saída do modelo
output = modelo(tokens["input_ids"], attention_mask=tokens["attention_mask"])
print("Forma da saída:", output.last_hidden_state.shape)
```

```python
# Testando o Mecanismo de Atenção com tokens de espaços vazios (Self-Attention)

frase2 = ["O mercado financeiro está estável",
         " "]

# Converta a frase em tokens:
tokens2 = tokenizer(frase2, padding=True, return_tensors="tf")
print(tokens2)
```

```python
# Obter a saída do modelo

print(tokens2["attention_mask"])
```

```python
# Resumos de textos.

from transformers import pipeline
resumo_modelo = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
texto = """ O mercado financeiro teve um crescimento significativo este ano, impulsionado por novos investimentos e
políticas econômicas favoráveis. As ações das empresas de tecnologia tiveram uma valorização expressiva,
atraindo mais investidores para o setor. Especialistas afirmam que a tendência deve continuar nos próximos meses,
apesar de desafios como a inflação e a taxa de juros elevada.
"""

resumo = resumo_modelo(texto, max_length=50, min_length=20, do_sample=False)
print("Resumo Gerado: ", resumo)
```
